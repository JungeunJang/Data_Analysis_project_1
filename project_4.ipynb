```python
import pandas as pd
import json
import folium

data = pd.read_csv('final_data_add3.csv', encoding='cp949')
data
```

```python
top5_1 = data[data['구별'] == '광진구']
top5_2 = data[data['구별'] == '마포구']
top5_3 = data[data['구별'] == '영등포구']
top5_4 = data[data['구별'] == '송파구']
top5_5 = data[data['구별'] == '종로구']

top5 = pd.concat([top5_1,top5_2,top5_3,top5_4,top5_5])
# top5.to_csv('data_top5.csv', encoding = 'utf-8')

top5
```

```python
borrow_t_data = top5.groupby(['구별', '사용년월'])['대여건수'].sum()
borrow_data = borrow_data.to_frame()
borrow_data.reset_index(inplace=True)

borrow_data['사용년월'] = pd.to_datetime(borrow_data['사용년월'])
borrow_data
```

```python
tail5_1 = data[data['구별'] == '양천구']
tail5_2 = data[data['구별'] == '용산구']
tail5_3 = data[data['구별'] == '금천구']
tail5_4 = data[data['구별'] == '강북구']
tail5_5 = data[data['구별'] == '도봉구']

tail5 = pd.concat([tail5_1,tail5_2,tail5_3,tail5_4,tail5_5])
tail5.to_csv('data_tail5.csv', encoding = 'utf-8')
```

```python
from datetime import datetime, date
import matplotlib.pyplot as plt
import seaborn as sns

# 한글문제 해결처리 
################################################################
import matplotlib.pyplot as plt 
from matplotlib import font_manager, rc
import platform
plt.rcParams['axes.unicode_minus'] = False
if platform.system() == 'Darwin':
    rc('font', family='AppleGothic')
elif platform.system() == 'Windows':
    path = "c:/Windows/Fonts/malgun.ttf"
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print('Unknown system... sorry~~~~')

borrow_top_data = top5.pivot_table('대여건수1', index = '사용년월', columns = '구별', aggfunc=sum).plot()
plt.title('상위 5구 대여건수')

borrow_tail_data = tail5.pivot_table('대여건수1', index='사용년월', columns='구별', aggfunc=sum).plot()
plt.title('하위 5구 대여건수')
```
